<b>Основні інструменти опису даних</b><br>
<b style="color:green">head(n)</b>: Повертає перші 'n' рядків з DataFrame. За замовчуванням 'n' дорівнює 5.<br>
<b style="color:green">tail(n)</b>: Повертає останні 'n' рядків з DataFrame. За замовчуванням 'n' дорівнює 5.<br>
<b style="color:green">describe()</b>: Повертає статистичний опис даних для кожного числового стовпця.<br>
<b style="color:green">info()</b>: Повертає інформацію про DataFrame, включаючи кількість рядків, тип даних кожного стовпця, кількість ненульових значень і споживання пам'яті.<br>
<b style="color:green">dtypes</b>: Повертає типи даних кожного стовпця.<br>
<b style="color:green">shape</b>: Повертає форму датафрейму як кортеж (кількість рядків, кількість стовпців).<br>
<b style="color:green">columns</b>: Повертає імена стовпців DataFrame.<br>
<b style="color:green">value_counts()</b>: Повертає серію, яка містить кількості унікальних значень.<br>
<b style="color:green">sort_values(by=column_name)</b>: Сортує DataFrame за значеннями в певному стовпці.<br>
<b style="color:green">corr()</b>: Повертає кореляційну матрицю для числових стовпців в DataFrame.<br>
<b style="color:green">isnull()</b>: Повертає DataFrame, де True позначає пропущені значення, а False - непропущені значення.<br>

```python
sets.merge(df, how = 'left', left_on = 'theme_id', right_on = 'id', suffixes = ['_theme', '_themes'])
```

```python
df = sets.merge(df, how = 'left', left_on = 'theme_id', right_on = 'id', suffixes = ['_theme', '_themes'])
```

```python
df[df['year'].isna()]
```

```python
# Перетворення року в строку
df['year'] = df['year'].astype(int).astype(str)

# Додавання дня і місяця до року
df['year'] = '01-01-' + df['year']

# Перетворення року в формат дати
df['year'] = pd.to_datetime(df['year'])
```

```python
df
```

```python
# Метод зведених таблиць - pd.pivot_table()
# https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html
grouped = pd.pivot_table(df[['year', 'name_themes']], index = ['year'], aggfunc = 'count')
```

```python
grouped
```

```python
grouped.reset_index(inplace = True)
```

```python
grouped
```

```python
grouped['year'] = grouped['year'].dt.year
```

```python
# стовпчикова діаграма: https://seaborn.pydata.org/generated/seaborn.barplot.html
sns.barplot(data=grouped, x="year", y="name_themes")
```

```python
sns.barplot(data=grouped, x="year", y="name_themes", errorbar = None)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

bar = sns.barplot(data=grouped, x="year", y="name_themes", errorbar = None)
bar.bar_label(bar.containers[0], fmt='%.1f')
bar.set_xticklabels(bar.get_xticklabels(), rotation=90)
```

```python
colors_df = pd.read_csv('lego/colors.csv')
inventory_parts_df =  pd.read_csv('lego/inventory_parts.csv')
inventories_df = pd.read_csv('lego/inventories.csv')
```

```python
df = df.merge(sets, how = 'left', on = 'set_num')
```

```python
df.info()
```

```python
df[['inventory_id']] = df[['inventory_id']].astype(str)
df[['part_num']] = df[['part_num']].astype(str)
```

```python
df[['inventory_id', 'part_num']].describe()
```

```python
df[['name_x', 'year']].describe()
```

```python
df[['name_x', 'year']]
df = df.groupby(['year', 'name_x']).size().reset_index(name='counts')
```

```python
df.sort_values(['year', 'counts'])
```

```python
df = df.sort_values(['year', 'counts'])
```

```python
sns.heatmap(df.pivot_table(index = 'name_x', columns = 'year', aggfunc = 'sum'))
```

```python
pivoted_df = df.pivot_table(index = 'name_x', columns = 'year', aggfunc = 'sum')
```

```python
pivoted_df
```

```python
# Розрахуємо суму по рядкам і додаємо її як нову колонку 'row_sum'
pivoted_df['row_sum'] = pivoted_df.sum(axis=1)

# Сортуємо мультиіндексний датафрейм по цій колонці в спадному порядку
pivoted_df= pivoted_df.sort_values('row_sum')

# Якщо вам не потрібна колонка 'row_sum', ви можете видалити її
pivoted_df= pivoted_df.drop(columns=['row_sum'])
```

```python
pivoted_df
```

```python
sns.set(rc={'figure.figsize':(11,20)})
sns.heatmap(pivoted_df)
```
